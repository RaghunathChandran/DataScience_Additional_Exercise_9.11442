# Data Exploration using Seaborn

```python
import matplotlib
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline 
```

```python
#Import the iris dataset:
from sklearn.datasets import load_iris
iris = load_iris()
headers = ['SepalLenghtCm','SepalWidthCm','PetalLenghtCm','PetalWidthCm','Species']
data = pd.DataFrame(iris.data, columns=iris.feature_names)
label = pd.DataFrame(list(map(lambda x: iris.target_names[x], iris.target)), columns=['Species'])
iris = pd.concat([data,label], axis=1 )
iris.head()
```

### 1. Use the distplot() to see the distribution of the SepalLenghtCm, SepalWidthCm, PetalLenghtCm, PetalWidthCm features. Plot them as subplots in a single image.

```python
fig, ax = plt.subplots(figsize=(20,7), ncols=2, nrows=2)
sns.distplot(iris['sepal length (cm)'], kde = False, ax=ax[0][0])
sns.distplot(iris['sepal width (cm)'],   kde = False, ax=ax[0][1])
sns.distplot(iris['petal length (cm)'],  kde = False, ax=ax[1][0])
sns.distplot(iris['petal width (cm)'],  kde = False, ax=ax[1][1])
```

### 2. Do a boxplot of all features except species.

```python
fig, ax = plt.subplots(figsize=(20,7), ncols=2, nrows=2)
sns.boxplot(iris['sepal length (cm)'], ax=ax[0][0])
sns.boxplot(iris['sepal width (cm)'], ax=ax[0][1])
sns.boxplot(iris['petal length (cm)'], ax=ax[1][0])
sns.boxplot(iris['petal width (cm)'], ax=ax[1][1])
```

### 3. Do a CountPlot for the feature 'Species'

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x="Species", data=iris)
```

### 4. Do a pairplot for the features 'SepalLenghtCm', 'SepalWidthCm', 'PetalLenghtCm', 'PetalWidthCm' and 'Species' 

```python
sns.pairplot(iris, height=2.5)
```

### 5. Do a lmplot for the features 'SepalLenghtCm', 'PetalLenghtCm'. Using hue display the different Species in different color.

```python
sns.lmplot(x="sepal length (cm)", y="petal length (cm)", data=iris, hue="Species")
```

### 6. Do a Bar plot for Species vs Sepal length

```python
sns.barplot(x="Species", y="sepal length (cm)", data=iris)
```

### 7. Using heatmap, plot the correlation matrix

```python
sns.heatmap(iris.corr())
```
